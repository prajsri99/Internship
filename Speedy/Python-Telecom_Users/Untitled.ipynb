```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("telecom_users.csv")
```

```python
data.head()
```

```python
data.describe
```

```python
data.isna()
```

```python
data_male_subscribers = data[data['gender']=='Male']
```

```python
data_non_senior = data_male_subscribers[data_male_subscribers['SeniorCitizen']==0]
```

```python
data_non_senior.head()
```

```python
data_filtered_subscribers = data_non_senior[
    (data_non_senior['PhoneService']=='Yes') &
    (data_non_senior['InternetService']!='No') &
    (data_non_senior['DeviceProtection']=='Yes') &
    (data_non_senior['StreamingTV']=='Yes') &
    (data_non_senior['PaperlessBilling']=='Yes')
]
data_filtered_subscribers
```

```python
total_male_subscribers = len(data_filtered_subscribers)
```

```python
print("Total Male Subscribers(Who are not senior citizens) with the specified services:",total_male_subscribers)
```

```python
data_female_subscribers = data[data['gender']=='Female']
```

```python
data_female_non_senior = data_female_subscribers[data_female_subscribers['SeniorCitizen']==0]
```

```python
data_female_non_senior.head()
```

```python
data_female_filtered_subscribers = data_female_non_senior[
    (data_female_non_senior['PhoneService']=='Yes') &
    (data_female_non_senior['InternetService']!='No') &
    (data_female_non_senior['DeviceProtection']=='Yes') &
    (data_female_non_senior['StreamingTV']=='Yes') &
    (data_female_non_senior['PaperlessBilling']=='Yes')
]
data_female_filtered_subscribers
```

```python
total_female_subscribers = len(data_female_filtered_subscribers)
```

```python
print("Total female Subscribers(Who are not senior citizens) with the specified services:",total_female_subscribers)
```
